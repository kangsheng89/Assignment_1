```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="ticks", color_codes=True)

from scipy.stats import zscore


from sklearn import model_selection
from sklearn.impute import SimpleImputer
```

```python
print('Jobs:\n', df['job'].unique())

df['job'] =df['job'].fillna("unknown")

print('Jobs:\n', df['job'].unique())

print('Marital:\n', df['marital'].unique())

df['marital'] =df['marital'].fillna("unknown")

print('Marital:\n', df['marital'].unique())

print('Education:\n', df['education'].unique())

df['education'] =df['education'].fillna("unknown")

print('Education:\n', df['education'].unique())

print('Default:\n', df['default'].unique())
print('Housing:\n', df['housing'].unique())
print('Loan:\n', df['loan'].unique())

print('Contact:\n', df['contact'].unique())

df['contact'] =df['contact'].fillna("unknown")

print('Contact:\n', df['contact'].unique())

print('Poutcome:\n', df['poutcome'].unique())

print('Credit History:\n', df['credit.history'].unique())
df['credit.history'] = df['credit.history'].str.replace('verygood','very good')
df['credit.history'] = df['credit.history'].str.replace('Poor','poor')
df['credit.history'] =df['credit.history'].fillna("unknown")
print('Credit History:\n', df['credit.history'].unique())

imp = SimpleImputer ()
df['age']=imp.fit_transform(df[['age']])
df['credit.score']=imp.fit_transform(df[['credit.score']])
df['work.exp']=imp.fit_transform(df[['work.exp']])

df.loc[df["month"] == "jan", "month_int"] = 1
df.loc[df["month"] == "feb", "month_int"] = 2
df.loc[df["month"] == "mar", "month_int"] = 3
df.loc[df["month"] == "apr", "month_int"] = 4
df.loc[df["month"] == "may", "month_int"] = 5
df.loc[df["month"] == "jun", "month_int"] = 6
df.loc[df["month"] == "jul", "month_int"] = 7
df.loc[df["month"] == "aug", "month_int"] = 8
df.loc[df["month"] == "sep", "month_int"] = 9
df.loc[df["month"] == "oct", "month_int"] = 10
df.loc[df["month"] == "nov", "month_int"] = 11
df.loc[df["month"] == "dec", "month_int"] = 12

df = df.drop(['month'],axis=1)
df = df.drop(['contact'],axis=1)

from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
df = pd.get_dummies(df, prefix_sep='_', drop_first=True)

df.head()
```

```python
corr = df.corr()
sns.set_context("notebook", font_scale=1.0, rc={"lines.linewidth": 2.5})
plt.figure(figsize=(20,20))
a = sns.heatmap(corr, annot=False, fmt='.2f')

rotx = a.set_xticklabels(a.get_xticklabels(), rotation=90)
roty = a.set_yticklabels(a.get_yticklabels())
```

```python
from sklearn.model_selection import train_test_split

import statsmodels.api as sm
y = df['Y']

X = df.drop(['Client Id','Y'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=100)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier(criterion = 'gini',max_depth=15)
dt_model.fit(X_train, y_train)
```

```python
y_pred = dt_model.predict(X_test)
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(confusion_matrix, annot=True,fmt='d')
```

```python
dt_score = dt_model.score(X_test, y_test)
dt_score
```

```python
print (dt_model.feature_importances_)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(criterion = 'entropy', max_depth = 18,min_samples_leaf=12)
rfcl = rfcl.fit(X_train, y_train)
y_pred = rfcl.predict(X_test)
```

```python
rfcl_score = rfcl.score(X_test, y_test)
rfcl_score
```

```python
from sklearn.ensemble import BaggingClassifier

bgcl = BaggingClassifier(n_estimators=50, max_samples= .7, bootstrap=True)
bgcl = bgcl.fit(X_train, y_train)
y_pred = bgcl.predict(X_test)
bgcl.score(X_test, y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
#abcl = AdaBoostClassifier(base_estimator=dt_model, n_estimators=50)
abcl = AdaBoostClassifier( base_estimator=bgcl, n_estimators= 50)
abcl = abcl.fit(X_train, y_train)
y_pred = abcl.predict(X_test)
abcl.score(X_test, y_test)
```

```python
from sklearn.neural_network import MLPClassifier

mlpc = MLPClassifier(hidden_layer_sizes=(15,5 ), 
                    max_iter=480, alpha=1e-5,
                    solver='lbfgs', verbose=10, 
                    tol=1e-5, random_state=1,
                    learning_rate_init=.0001)
mlpc = mlpc.fit(X_train, y_train)
y_pred = mlpc.predict(X_test)
mlpc.score(X_test, y_test)
```
