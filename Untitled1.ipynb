```python
def score_model(model):
    # Fit the model on train
    model.fit(X_train, y_train)
    #predict on test
    y_predict = model.predict(X_test)
    score = cross_val_score(model, X_train, y_train, cv=5)
    return score,y_predict

def draw_cm(actual,predicted):
    cm = metrics.confusion_matrix(actual,predicted)
    sns.heatmap(cm, annot=True,  fmt='d')
    plt.show()
```

```python
all_score ={}
for k,v in models_list.items():
    score,y_predict = score_model(v)
    print ('\nModel %s:' % (k))
    print ('\tAccuracy: %0.2f (+/- %0.2f)'% ( score.mean(), score.std() * 2))
    print ('\tROC_AUC_score: %0.2f'% (roc_auc_score(y_test, y_predict)))
    print ('\tF1_score: %0.2f'% (f1_score(y_test, y_predict)))
    print ('\nClassification_Report:\n'+classification_report(y_test, y_predict))
    all_score.update({k:(score.mean(),score.std(),f1_score(y_test, y_predict),roc_auc_score(y_test, y_predict))})
    draw_cm(y_test,y_predict)
```

```python
feat_imp_dict = dict(zip(X.columns, grid_search_rf.best_estimator_.feature_importances_))    
feat_imp = pd.DataFrame.from_dict(feat_imp_dict, orient='index')
print (feat_imp.sort_values(by=0, ascending=False))
plt.figure(figsize=(10,8))
sns.barplot(x=feat_imp[0],y=feat_imp.index)
```
